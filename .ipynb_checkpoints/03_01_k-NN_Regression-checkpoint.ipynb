# k-NN Regression k-최근접 이웃 회귀

## 지도 학습 알고리즘 (supervised learning algorithm)

1. 분류(classification): sample을 몇 개의 class중 하나로 분류하는 방법
    
2. 회귀(regression): 두 변수 사이의 상관관계를 분석해서 임의의 어떤 숫자를 예측하는 방법

## k-최근접 이웃 회귀 (k-NN Regression)

- 예측하려는 sample에 가장 가까운 샘플 k개를 선택

    - 이웃한 sample의 target은 class가 아니라 임의의 수치 $\rightarrow$ 수치들의 평균으로 예측

## Data Collection

```python
import numpy as np

perch_length = np.array([8.4, 13.7, 15.0, 16.2, 17.4, 18.0, 18.7, 19.0, 19.6, 20.0, 21.0,
       21.0, 21.0, 21.3, 22.0, 22.0, 22.0, 22.0, 22.0, 22.5, 22.5, 22.7,
       23.0, 23.5, 24.0, 24.0, 24.6, 25.0, 25.6, 26.5, 27.3, 27.5, 27.5,
       27.5, 28.0, 28.7, 30.0, 32.8, 34.5, 35.0, 36.5, 36.0, 37.0, 37.0,
       39.0, 39.0, 39.0, 40.0, 40.0, 40.0, 40.0, 42.0, 43.0, 43.0, 43.5,
       44.0])

perch_weight = np.array([5.9, 32.0, 40.0, 51.5, 70.0, 100.0, 78.0, 80.0, 85.0, 85.0, 110.0,
       115.0, 125.0, 130.0, 120.0, 120.0, 130.0, 135.0, 110.0, 130.0,
       150.0, 145.0, 150.0, 170.0, 225.0, 145.0, 188.0, 180.0, 197.0,
       218.0, 300.0, 260.0, 265.0, 250.0, 250.0, 300.0, 320.0, 514.0,
       556.0, 840.0, 685.0, 700.0, 700.0, 690.0, 900.0, 650.0, 820.0,
       850.0, 900.0, 1015.0, 820.0, 1100.0, 1000.0, 1100.0, 1000.0,
       1000.0])
```

### scatter plot

```python
import matplotlib.pyplot as plt

plt.scatter(perch_length, perch_weight)
plt.xlabel('length')
plt.ylabel('weight')

plt.show()
```

## Data Preprocessing

```python
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(
    perch_length, perch_weight, random_state=42
)
```

- scikit-learn의 train set: 2차원 배열

   -  `perch_length`: 1차원 배열 $\rightarrow$ `train_input`, `test_input`: 1차원 배열 $\rightarrow$ 2차원으로 변형

<br>

- `numpy` method

    - `ndarray.reshape(shape)`: 바꾸려는 배열의 크기`shape` 지정

```python
train_input = train_input.reshape(-1, 1)
test_input = test_input.reshape(-1, 1)

train_input.shape,  test_input.shape
```

## Training Model

```python
from sklearn.neighbors import KNeighborsRegressor

knr = KNeighborsRegressor()

knr.fit(train_input, train_target)
```

## Evaluating Model

```python
knr.score(test_input, test_target)
```

### 결정계수 $R^2$

$$ R^2 = 1 - \frac{\sum{{(target - predict)^2}}}{\sum{{(target - mean)^2}}} $$

- `knr = KNeighborsRegressor()`$\rightarrow$ `knr.score()` = 결정계수

- scikit-learn의 `score()` method $\rightarrow$ output value: 높을수록 좋은 값

```python
from sklearn.metrics import mean_absolute_error

test_prediction = knr.predict(test_input)

mae = mean_absolute_error(test_target, test_prediction)
mae
```

- `mean_abosolute_error`: target과 predcition의 절댓값 오차의 평균
    
    - `19.157...`: 예측이 평균적으로 target과 약 19g 차이난다

## 과대적합 vs. 과소적합 <br> Overfitting vs. Underfitting

```python
knr.score(train_input, train_target)
```

- generallly, train set에 fit $\rightarrow$ train set에 잘 맞는 모델 $\rightarrow$ train set의 점수가 높다.

    - model이 train set에 **과대적합, overfitting**되었다: train set의 점수 >> test set의 점수
    
    - model이 train set에 **과소적합, underfitting**되었다: train set의 점수가 더 나쁘거나, 두 점수 모두 나쁘다.
    
        - underfitting: model이 너무 단순해서 train set에 적절하게 훈련되지 않은 경우, 부족하게 학습된 경우

<br>

- model `knr`: test set score > train set score $\rightarrow$ underfitting 발생
  
    - 모델을 더 복잡하게 만들어서 해결
    
<br>

- k-NN algorithm: 이웃의 수 `k`

    - `k`$\downarrow$ : local pattern에 민감해진다 $\rightarrow$ model이 더 복잡해진다
    
    - `k`$\uparrow$ : global pattern에 따른다 $\rightarrow$ model이 더 단순해진다

```python
knr.n_neighbors = 3

knr.fit(train_input, train_target)
knr.score(train_input, train_target)
```

```python
knr.score(test_input, test_target)
```

- test set score $<$ train set score $\rightarrow$ underfitting 해결

- test set score $\simeq$ train set score $\rightarrow$ overfitting 해결

- - -

```python
knr = KNeighborsRegressor()

x = np.arange(5, 45).reshape(-1, 1)

for k in [1, 5, 10]:
    knr.n_neighbors = k
    knr.fit(train_input, train_target)
    
    prediction = knr.predict(x)
    
    plt.scatter(train_input, train_target)
    plt.plot(x, prediction)
    
    plt.title(f'n_neibors = {k}')
    plt.xlabel('length')
    plt.ylabel('weight')
    
    plt.show()
```

- k-NN regression에서 `n_neighbors` parameter가 증가할수록 model은 단순해진다: 예측이 선형적(linear)이다
